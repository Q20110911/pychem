# Read data

```python
def GetMolsForAssay(aid):
    """ pulls molecules from ChEMBL for a particular assay ID
      The molecules "_Name" property is set to molregno
      and the target for the assay is stored in the "ChEMBL_Target" property
    """
    data = %sql postgresql://localhost/chembl_16 \
          select pref_name from assays join target_dictionary using (tid) where assay_id=:aid ;
    target=str(data[0][0])
    data = %sql postgresql://localhost/chembl_16 \
        select molregno,canonical_smiles smiles from \
          (select distinct molregno from \
            activities where assay_id=:aid ) tbl \
          join compound_structures using (molregno) ;
    ms =[]
    for molregno,smi in data:
        m = Chem.MolFromSmiles(smi)
        if not m: 
            continue
        m.SetProp("_Name",str(molregno))
        m.SetProp("ChEMBLTarget",target)
        ms.append(m)
    print("Assay ID: {0}, Target: {1}, {2} molecules".format(aid,target,len(ms)))

    return ms
```

```python
def MCS_Report(ms,**kwargs):
    mcs = MCS.FindMCS(ms,**kwargs)
    nAts = numpy.array([x.GetNumAtoms() for x in ms])
    print nAts.mean(),mcs.numAtoms
    print mcs.smarts
    mcsM = Chem.MolFromSmarts(mcs.smarts)
    mcsM.UpdatePropertyCache()
    Chem.SetHybridization(mcsM)
    img=Draw.MolToImage(mcsM,kekulize=False)
    return mcsM,img
```

```python
mcsM,img = MCS_Report(ms)
img
```

```python
mcsM,img = MCS_Report(ms,atomCompare="any",threshold=0.8)
img
```

```python
mcsM,img = MCS_Report(ms,threshold=0.8)
img
```

```python
def AlignMols(ms,mcsM):
    cms=[]
    matches=[]
    for m in ms:
        cm = Chem.Mol(m)
        match=cm.GetSubstructMatch(mcsM)
        if match:
            AllChem.GenerateDepictionMatching2DStructure(cm,mcsM)
            Chem.Kekulize(cm)
            cms.append(cm)
            matches.append(match)
    print('MCS matches {0} of {1} molecules'.format(len(cms),len(ms)))
    img=Draw.MolsToGridImage(cms,molsPerRow=5,kekulize=False,highlightAtomLists=matches,legends=[m.GetProp("_Name") for m in cms])
    return cms,img
```

```python
cms,img=AlignMols(ms,mcsM)
img
```

```python
ms = GetMolsForAssay(563770)
mcsM,img = MCS_Report(ms,threshold=0.8,atomCompare='any',completeRingsOnly=True,ringMatchesRingOnly=True)
cms,img=AlignMols(ms,mcsM)
img
```
